# Packages and Modules
A module in Python is a file containing Python code (functions, variables, or classes) that can be imported and reused in another program.

## import methods
- `import math` : Access via math.sqrt() <br><br>
- `from math import *` : Import all (not recommended)<br><br>
- `from math import sqrt` : Access directly: sqrt()<br><br>
- `import math as m` : Alias: m.pi, m.sqrt()<br><br>


```python
import datetime
datetime_object = datetime.datetime.now()
print(datetime_object)

datetime_object2 = datetime.date.today()
print(datetime_object2)
```

```python
import datetime

# Get current time with local timezone info
local_tz = datetime.datetime.now().astimezone().tzinfo

print(f"Timezone Name: {local_tz.tzname(None)}") 
print(f"Timezone Detail: {local_tz}")             
```

```python
import math

print(math.pi)
```

```python
import math as m
print(m.pi)
```

```python
import statistics as st
data = [x for x in range(1, 15)]
print(data)
mean_value = st.mean(data)
print(mean_value)
```
