# Scope in Python
Scope refers to the coding from which a particular Python object is accessible. Hence, once cannot access any particular object from anywhere from the code, the accessing has to be allowed by the scope of the object.

Types of scopes:
1. Local
2. Enclosing
3. Global
4. Built-in

## Local Scope
Veriable defined inside a fuction, accessible only within that function.
## Enclosing Scope
Variables in an outer function, accessible in nested (inner) functions.
## Global Scope

```
Python follows LEGB rule


````python
x = "global"

def outer():
    x = "enclosing"

    def inner():
        x = "local"
        print(x)    # Local
        
    inner()
    print(x)    # Enclosing

outer()
print(x)        # Global
````
